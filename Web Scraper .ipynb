```python
# Merge all game stats
games=[df,df1,df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11,df12,df13,df14,df15]
all_games = pd.concat(games, ignore_index=True, sort=False)
all_games
```

```python
all_games['Week'] = 'WK1'
all_games.head()
```

```python
all_games = all_games.set_index('Player')
all_games.head()
```

```python
all_games.columns
```

```python
all_games = all_games.rename(columns={"Yds.1": "Yds"})
all_games.head()
```

```python
cols=pd.Series(all_games.columns)
for dup in all_games.columns[all_games.columns.duplicated(keep=False)]: 
    cols[all_games.columns.get_loc(dup)] = ([dup + '.' + str(d_idx) 
                                     if d_idx != 0 
                                     else dup 
                                     for d_idx in range(all_games.columns.get_loc(dup).sum())]
                                    )
all_games.columns=cols
```

```python
all_games = all_games.fillna(0)
all_games.head()
```

```python
# Saving as a csv 
all_games.to_csv('Resources/Week_1.csv')
```
