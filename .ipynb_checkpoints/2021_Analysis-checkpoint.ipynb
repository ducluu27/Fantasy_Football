```python
df_2018 = pd.read_csv('Resources/FF_Stats_2018.csv')
df_2018.head()
```

```python
df_2018 = df_2018.drop(columns=['Rk', 'Tm'])
df_2018.head()
```

```python
df_2018 = df_2018.set_index('Player')
df_2018.head()
```

```python
df_2020.columns
```

```python
df_2020.dtypes
```

```python
all_df = [df_2020,df_2019,df_2018]
merge_df = pd.concat(all_df)
merge_df.head()
```

```python
merge_df.count()
```

```python
new_merge_df = merge_df.fillna(0)
new_merge_df.head()
```

```python
new_merge_df=new_merge_df.sort_index(ascending=True)
new_merge_df.head()
```

```python
player_df = new_merge_df.groupby('Player').std()
player_df.head()
```

```python
new_player_df = player_df.dropna()
new_player_df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
from matplotlib import pyplot 
```

```python
# Split data into training and testing
X = new_player_df[['Age', 'G', 'GS', 'Cmp', 'Att', 'Yds', 'TD', 'Int', 'Att.1',
       'Yds.1', 'Y/A', 'TD.1', 'Tgt', 'Rec', 'Yds.2', 'Y/R', 'TD.2', 'Fmb',
       'FL', 'TD.3', '2PM', '2PP', 'FantPt', 'PPR', 'DKPt', 'FDPt', 'VBD',
       'PosRank']]

y = new_player_df['OvRank']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=100)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
print(y_predict)
```

```python
# Feature importance
# fit the model
model.fit(X_train, y_train)
# get importance
importance = model.coef_
# summarize feature importance
for i,v in enumerate(importance):
	print ('%30s, Score: %.5f' % (X_train.columns[i],v))
# plot feature importance
pyplot.bar([x for x in range(len(importance))], importance)
pyplot.show()
```

```python
# Feature Importance into a dataframe
important_features = model.coef_
columns=X_train.columns
feature_import_df = pd.DataFrame({"feature":columns,"importance":important_features})
new_feature_df = feature_import_df.sort_values(by='importance',ascending=False)
new_feature_df = new_feature_df.round(5)
new_feature_df
```
