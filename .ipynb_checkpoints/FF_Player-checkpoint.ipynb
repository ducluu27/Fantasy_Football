```python
# Resources
import numpy as np
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

```python
#df_2020 = pd.read_csv('Resources/FF_Stats_2020.csv')
#df_2020.head()
URL = "https://www.pro-football-reference.com/years/2020/fantasy.htm"
df_2020 = pd.read_html(URL, match = 'Fantasy Rankings')
```

```python
#converting table into pandas dataframe and droping extra header column
df_2020[0]
df_2020_data = pd.DataFrame(data=df_2020[0])
df_2020_data.columns = df_2020_data.columns.droplevel(0)
df_2020_data.head()
```

```python
df_2020_data["Year"] = '2020'
df_2020_data.head()
```

```python
df_2020_data.columns
```

```python
df_2020_data = df_2020_data[['Rk', 'Player', 'Tm', 'Year', 'FantPos', 'Age', 'G', 'GS', 'Cmp', 'Att', 'Yds',
       'TD', 'Int', 'Att', 'Yds', 'Y/A', 'TD', 'Tgt', 'Rec', 'Yds', 'Y/R',
       'TD', 'Fmb', 'FL', 'TD', '2PM', '2PP', 'FantPt', 'PPR', 'DKPt', 'FDPt',
       'VBD', 'PosRank', 'OvRank']]
df_2020_data.head()
```

```python
spec_chars = ["!",'"',"#","%","&","'","(",")",
              "*","+",",","-",".","/",":",";","<",
              "=",">","?","@","[","\\","]","^","_",
              "`","{","|","}","~","–"]
for char in spec_chars:
    df_2020_data['Player'] = df_2020_data['Player'].str.replace(char, ' ')
```

```python
df_2020_data.head()
```
